```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
ciiu = pd.read_csv("./data/input/Estructura_empresarial_Medell_n_seg_n_comunas_y_actividad_econ_mica_20241110.csv")
ciiu.head()
```

```python
df_filtered = ciiu[ciiu["AÑO"] == 2022]
df_filtered.head()
```

```python
df_filtered.describe()
```

```python
ciius = [5511, 5512, 5513, 5514, 5519, 7911, 7912, 7990, 5611, 5613, 9321, 9329]
```

```python
df_final = df_filtered.copy()
df_final = df_final[df_final["CIIU"].isin(ciius)]
df_final.loc[df_final["CIIU"].isin([5511, 5512, 5513, 5514, 5519]), "Categoría"] = "Alojamiento"
df_final.loc[df_final["CIIU"].isin([7911, 7912, 7990]), "Categoría"] = "Turismo"
df_final.loc[df_final["CIIU"].isin([5611, 5613]), "Categoría"] = "Gatronomía"
df_final.loc[df_final["CIIU"].isin([9321, 9329]), "Categoría"] = "Recreación"
df_final.head()
```

```python
df_final.columns
```

```python
places = df_final.drop(["AÑO", "CIIU", "Descripción", "Categoría"], axis=1).columns

df_grouped = df_final.groupby("Categoría").sum()[places].reset_index()
df_grouped.head()
```

```python
for i in range(len(df_grouped)):
    print()

    values = df_grouped.iloc[i].values
    values_order, places_order = zip(*sorted(zip(values[1:], places), reverse=True))

    plt.figure(figsize=(12, 6))
    plt.title(values[0])


    sns.barplot(x = values_order, y = places_order,
                hue = places_order ,palette = "coolwarm"
                )
    plt.ylabel('Comuna')
    plt.xlabel('Cantidad de establecimientos')
    plt.tight_layout()
    plt.show()
```

```python
df_sum = df_grouped.sum()[1:].reset_index()
df_sum.columns = ['Comuna', 'Valor']

df_sum['Comuna'] = df_sum['Comuna'].astype(str)
df_sum['Valor'] = df_sum['Valor'].astype(int)

df_sum.sort_values("Valor", inplace=True, ascending=False)

top_3 = df_sum.head(3)
otros = df_sum.tail(df_sum.shape[0] - 3)["Valor"].mean()

otros_data = pd.DataFrame({'Comuna': ['Otros'], 'Valor': otros})

df_otros_final = pd.concat([top_3, otros_data])


plt.figure(figsize=(12, 6))
sns.barplot(x='Comuna', y='Valor', data=df_otros_final, hue="Valor", palette="coolwarm")

# Personalizar el gráfico
plt.title('Cantidad de Establecimientos por Comuna')
plt.xlabel('Comuna')
plt.ylabel('Cantidad de Establecimientos')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
