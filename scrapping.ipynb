```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
import re
import time
import datetime
```

```python
url = "https://www.google.com/maps/@6.2297192,-75.5601012,13.19z?hl=es&entry=ttu&g_ep=EgoyMDI0MTEwNi4wIKXMDSoASAFQAw%3D%3D"
```

```python
driver = webdriver.Chrome()
```

```python
data = {
    "name": [],
    "desc": [],
    "score": [],
    "c_score": [],
    "price": [],
    "category": [],
    "accessibility": [],
    "schedule": [],
    "web": [],
    "search_parameters": [],
    "phone": [],
    "address": [],
    "lat": [], 
    "lon": []
}

for place in places["Medellín"]["Comunidades"]:
    for site in sites:

        driver.get(cords[place])

        search_input = driver.find_element(By.XPATH, '//*[@id="searchboxinput"]')
        search_button = driver.find_element(By.XPATH, '//*[@id="searchbox-searchbutton"]/span')
        
        search_input.clear()
        
        search_input.send_keys(f"{site} en {place}")
        search_button.click()

        driver.implicitly_wait(4)

        sites_lists = driver.find_elements(By.CLASS_NAME, 'hfpxzc')
        if site not in common:
            driver.execute_script("arguments[0].scrollIntoView()", sites_lists[-1])
            sites_lists = driver.find_elements(By.CLASS_NAME, 'hfpxzc')

        publicity =  len(driver.find_elements(By.XPATH, "//*[@class='geckTe FcJlLe']"))
        if publicity != 0:
            sites_lists = sites_lists[publicity:]

        for i in sites_lists:
            driver.execute_script("arguments[0].scrollIntoView()", i)
            time.sleep(0.5)
            url = i.get_attribute("href")
            driver.execute_script(f"window.open('{url}', '_blank');")
            driver.switch_to.window(driver.window_handles[1])
            try:
                name = driver.find_elements(By.XPATH, "//*[@class='DUwDvf lfPIob']")
                if len(name) > 0:
                    
                    data["name"].append(name[0].text)
                else:
                    data["name"].append(pd.NA)

                scores = driver.find_elements(By.XPATH, "//*[@class='F7nice ']")
                if len(scores) > 0:
                    score = pd.NA
                    c_score = pd.NA
                    try:
                        score, c_score = tuple(map(lambda x: float(x), scores[0].text.replace("(", "").replace(")", "").replace(".", "").replace(",", ".").split("\n")))
                    except:
                        pass
                    
                    data["score"].append(score)
                    data["c_score"].append(c_score)
                else:
                    data["score"].append(pd.NA)
                    data["c_score"].append(pd.NA)

                prices = driver.find_elements(By.XPATH, "//*[@class='mgr77e']")
                if len(prices) > 0:
                    
                    data["price"].append(prices[0].text.replace("·", ""))
                else:
                    data["price"].append(pd.NA)
                    
                
                category = driver.find_elements(By.XPATH, "//*[@class='DkEaL ']")
                if len(category) > 0:
                    
                    data["category"].append(category[0].text)
                else:
                    data["category"].append(pd.NA)
                    
                accessibility = driver.find_elements(By.XPATH, "//*[@class='wmQCje google-symbols']")
                if len(accessibility) > 0:
                    
                    data["accessibility"].append(accessibility[0].get_attribute("aria-label"))
                else:
                    data["accessibility"].append(pd.NA)

                webs = driver.find_elements(By.XPATH, "//*[@class='CsEnBe']")
                
                driver.execute_script("arguments[0].scrollIntoView()", webs[0])
                
                
                address = driver.find_elements(By.XPATH, "//*[@class='Io6YTe fontBodyMedium kR99db fdkmkc ']")
                if len(address) > 0:
                    
                    data["address"].append(address[0].text)
                else:
                    data["address"].append(pd.NA)
                    
                if len(webs) > 0:
                    for web in webs:
                        if web.get_attribute("aria-label") is not None:
                            if web.get_attribute("aria-label").startswith("Teléfono: "):
                                
                                data["phone"].append(web.get_attribute("aria-label").replace("Teléfono: ", ""))
                            elif web.get_attribute("aria-label").startswith("Sitio web: "):
                                if web.get_attribute("href") is not None and web.get_attribute("href") != "":
                                    
                                    data["web"].append(web.get_attribute("href"))
                                else:
                                    data["web"].append(pd.NA)
                else:
                    data["phone"].append(pd.NA)
                    data["web"].append(pd.NA)
                    
                if len(data["web"]) != len(data["name"]):
                    data["web"].append(pd.NA)
                if len(data["phone"]) != len(data["name"]):
                    data["phone"].append(pd.NA)
                
                
                data["search_parameters"].append(f"{place} {site}")
                schedules = driver.find_elements(By.XPATH, "//*[@class='g2BVhd eoFzo ']")
                if len(schedules) > 0:
                    driver.execute_script("arguments[0].scrollIntoView()", schedules[0])
                    time.sleep(1)
                    
                    inicio = None
                    fin = None
                    try:
                        for horario in driver.find_elements(By.XPATH, "//*[@class='dpoVLd finExf']"):
                            # Extraemos la ocupación y la hora del string
                            ocupacion = int(re.findall("ocupación: (\d*)", horario.get_attribute("aria-label"))[0])
                            hora = horario.get_attribute("aria-label").split("hora:")[1].strip().replace(' ', ' ')

                            if ocupacion > 0 and inicio is None:
                                # Establecer el inicio del rango cuando la ocupación no es 0
                                inicio = hora
                            if ocupacion == 0 and inicio is not None:
                                # Establecer el final del rango cuando la ocupación vuelva a ser 0
                                fin = hora
                                break

                        data["schedule"].append(str(str(inicio) + "," + str(fin)).replace("p.\xa0m.)", "pm").replace("a.\xa0m.)", ""))
                    except:
                        data["schedule"].append(pd.NA)
                else:
                    data["schedule"].append(pd.NA)
                
                driver.execute_script("arguments[0].scrollIntoView()", name[0])
                time.sleep(1)
                
                info_button = driver.find_elements(By.XPATH, "//*[@class='hh2c6 ']")
                try:
                    info_button[-1].click()
                    info = driver.find_elements(By.XPATH, "//*[@class='m6QErb DxyBCb kA9KIf dS8AEf XiKgde ']")

                    if len(info) > 0:
                        data["desc"].append(str(info_button[-1].get_attribute("aria-label").split()[0] + ",".join([i.text for i in info])))
                    else:
                        data["desc"].append(pd.NA)
                except:
                    data["desc"].append(pd.NA)
                

                time.sleep(3)
                lat, lon = [float(x) for x in re.findall(r"@(.*),", driver.current_url)[0].split(",")]
                data["lat"].append(lat)
                data["lon"].append(lon)
            except:
                max = 0
                for i, list in data.items():
                    if max < len(list):
                        max = len(list)
                        
                for i , list in data.items():
                    if len(list) > max:
                        list.pop()

                data["name"].append(pd.NA)
                data["score"].append(pd.NA)
                data["c_score"].append(pd.NA)
                data["price"].append(pd.NA)
                data["category"].append(pd.NA)
                data["accessibility"].append(pd.NA)
                data["address"].append(pd.NA)
                data["phone"].append(pd.NA)
                data["web"].append(pd.NA)
                data["schedule"].append(pd.NA)
                data["desc"].append(pd.NA)
                data["lat"].append(pd.NA)
                data["lon"].append(pd.NA)
                data["search_parameters"].append(pd.NA)
                
            driver.close()

            driver.switch_to.window(driver.window_handles[0])
```
